```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("globalWarming.csv")
data
```

```python
plt.scatter(data["date"], data["LandAverageTemperature"], s = 50, c = "red")
plt.xlabel("date")
plt.ylabel("temperature")
plt.show()
```

```python
data["LandAverageTemperature"] = (data["LandAverageTemperature"] - data["LandAverageTemperature"].mean())/data["LandAverageTemperature"].std()
```

```python
data["LandAverageTemperature"]
```

```python
plt.scatter(data["LandAverageTemperature"], data["date"], s = 50, c = "red")
plt.xlabel("date")
plt.ylabel("temperature")
plt.show()
```

```python
p10 = np.poly1d(np.polyfit(x = data["date"], y = data["LandAverageTemperature"], deg = 10))
p8 = np.poly1d(np.polyfit(x = data["date"], y = data["LandAverageTemperature"], deg = 15))
```
