```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("globalWarming.csv")
data
```

```python
plt.scatter(data["date"], data["LandAverageTemperature"], s = 50, c = "red")
plt.xlabel("date")
plt.ylabel("temperature")
plt.show()
```

```python
p10 = np.poly1d(np.polyfit(x = data["date"], y = data["LandAverageTemperature"], deg = 10))
p8 = np.poly1d(np.polyfit(x = data["date"], y = data["LandAverageTemperature"], deg = 15))
```

```python
xp = np.linspace(1850, 2016, 165)
```

```python
plt.plot(data['date'], data['LandAverageTemperature'], 'o', xp, p10(xp), '-', xp, p8(xp), '--')
```

```python
X = np.linspace(1850, 2016, 165)
```

```python
plt.plot(X, p10(X))
plt.show()
```
